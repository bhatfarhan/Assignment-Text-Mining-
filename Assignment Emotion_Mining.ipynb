```python
#Sentiment Analysis
afinn = pd.read_csv("C:\\Users\\farhanfarooq\\Downloads\\Afinn.csv",
                   sep=',',encoding='latin-1')
afinn.shape
```

```python
elon = pd.read_csv("C:\\Users\\farhanfarooq\\Desktop\\work files\\ExcelR\\Assignments\\Assign. Text mining\\Elon_musk Text Mining.csv")
elon
```

```python
elon = [Text.strip() for Text in elon.Text]
elon = [Text for Text in elon if Text]
elon[0:10]
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk import tokenize
sentences = tokenize.sent_tokenize(''.join(elon))
sentences[5:15]
```

```python
sent_df = pd.DataFrame(sentences, columns=['Sentence'])
sent_df
```

```python
affinity_scores = afinn.set_index('word')['value'].to_dict()
```

```python
affinity_scores
```

```python
nlp = spacy.load('en_core_web_md')
sentiment_lexicon = affinity_scores

def calculate_sentiment(text:str = None):
    sent_score = 0
    if text:
        sentence = nlp(text)
        for word in sentence:
            sent_score += sentiment_lexicon.get(word.lemma_,0)
    return sent_score
```

```python
#Test whether it works
calculate_sentiment(text = '')
```

```python
sent_df['sentiment_value'] = sent_df["Sentence"].apply(calculate_sentiment)
```

```python
#how many words are in the sentence
sent_df['word_count'] = sent_df['Sentence'].str.split().apply(len)
sent_df['word_count'].head(10)
```
